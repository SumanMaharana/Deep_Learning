```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import load_iris
```

```python
data = load_iris()
data
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df.head()
```

```python
df2 = pd.DataFrame(data.target, columns = ["Target"])
df2.sample(5)
```

```python
## encode class values as integers
# encoder = LabelEncoder()
# encoder.fit(Y)
# encoded_Y = encoder.transform(Y)

from keras.utils import np_utils
encoded_y = np_utils.to_categorical(df2)
encoded_y
```

```python
model = Sequential()
model.add(Dense(8, input_dim=4, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
#adam = mini batch gradient descent algo
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics="accuracy")
```

```python
model.fit(df, encoded_y, epochs=100, batch_size=10)
```

```python
predictions = model.predict(df)
```

```python
predictions
```

```python
for i in range(35,130,3):
  print(predictions[i], encoded_y[i])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df,encoded_y,test_size=0.25,random_state=2)
```

```python
y_pred = model.predict(x_test)

y_pred
```

```python
def vec2scalar(list_):
    pred = []
    for i in list_:
        val = [round(j,0) for j in i]
        if val[0]==1.0:
            pred.append(0)
        elif val[1]==1.0:
            pred.append(1)
        else:
            pred.append(2)
    return pred


print(vec2scalar(y_pred))

```

```python
print(vec2scalar(y_test))
```

```python
_, accuracy = model.evaluate(df, encoded_y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
